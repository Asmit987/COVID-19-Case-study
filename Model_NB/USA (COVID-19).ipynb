```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot, plot
import plotly.offline as pyo
pyo.init_notebook_mode()
```

```python
df_confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")


df_death = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv")
```

```python
df_confirmed['Country_Region'] = 'USA'
df_confirmed.rename(columns = {'Province_State':'State','Country_Region':'Country'}, inplace = True)
df = df_confirmed.drop(columns = ['UID','iso2','iso3','code3','FIPS', 'Admin2','Country','Lat','Long_','Combined_Key'], axis = 1)
count = df.groupby(['State']).sum().sort_values(['7/4/20'], ascending= True)
confirmed = count.sort_values(['7/6/20'],ascending = False).transpose()
```

```python
df_death['Country_Region'] = 'USA'
df_death.rename(columns = {'Province_State':'State','Country_Region':'Country'}, inplace = True)
dfd = df_death.drop(columns = ['UID','iso2','iso3','code3','FIPS', 'Admin2','Country','Lat','Long_','Combined_Key'], axis = 1)
count_d = dfd.groupby(['State']).sum().sort_values(['7/4/20'], ascending= True)
death = count_d.sort_values(['7/6/20'],ascending = False).transpose()
```
